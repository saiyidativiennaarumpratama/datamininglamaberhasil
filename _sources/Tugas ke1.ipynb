1. Membaca data dari URL

```python
import pandas as pd
url='https://gist.githubusercontent.com/meperezcuello/82a9f1c1c473d6585e750ad2e3c05a41/raw/d42d226d0dd64e7f5395a0eec1b9190a10edbc03/Medical_Cost.csv'
c=pd.read_csv(url)
print(c)
```

```python
harga =c['charges']
harga
```

2. Membuat statistik diskriptif dari data (setiap atribut yang bertipe numerik) 

```python
import statistics as st
```

```python
st.median(harga)
```

```python
st.mode(harga)
```

```python
st.mean(harga)
```

```python
harga.skew()
```

3. Mendiskripsikan kemiringan/skewned dari data setiap variabel (bertipe numerik)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
sns.distplot(harga)
plt.show()
```
